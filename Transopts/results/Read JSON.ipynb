```python
import json
import os
import matplotlib.pyplot as plt
```

```python
names = os.listdir()
print(names)
```

```python
file = open(names[-1])
```

```python
json_obj = json.load(file)
```

```python
models = list(json_obj.keys())
print(models)
```

```python
ite = 0
fig, ax = plt.subplots(nrows=2, ncols=5, figsize=(20,6), squeeze = False )
for row in ax:
    for col in row:
        col.plot(json_obj[models[ite]]['losses'])
        col.set_title(models[ite])
        col.set_ylim([0, 4])
        ite += 1

fig.subplots_adjust(hspace=0.3, wspace=0.4)        
fig.suptitle('Loss Curves')
plt.savefig("Loss Curves/curves.jpg")
plt.show()
```
